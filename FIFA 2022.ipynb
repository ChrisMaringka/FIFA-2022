```python
analysis1 = df.loc[(df['club_name'].str.contains('Manchester|Liverpool', regex= True, na=False)) ]
```

```python
print(analysis1)
```

```python
analysis1.boxplot('overall', by='club_name', widths=(0.5,0.5,0.5), return_type='axes')
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('overall ~ club_name', data=analysis1).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```
